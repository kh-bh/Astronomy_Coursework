```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

#### Problem 1

```python
datasite='http://burro.case.edu/Academics/Astr222/HW/HW1/'
datafile='NmAPQ3.dat'
```

```python
# this next line loads the data into an astropy table, telling astropy
# that the data starts at the first non-commented line, and the column names
# start two lines before that
starcounts=ascii.read(datasite+datafile,header_start=-2,data_start=0)
print(starcounts.colnames) # will list the column names
```

```python
# Store each of the columns
mag=starcounts['mag']
logN_up=starcounts['logN_up']
logN_in=starcounts['logN_in']
```

```python
# A plot showing logN(m) vs m in each direction on the same plot as well as a uniform distribution
plt.scatter(mag, logN_up, color = 'orange', alpha = .7)
plt.scatter(mag, logN_in, color = '#88c999', alpha = .7)
plt.plot(mag, .6*mag - 5, color = 'darkblue')
plt.legend(["Up Out of Galactic Plane" , "In towards Galactic Center", "Uniform Behavior (C = -5)"])
plt.xlabel("Magnitude")
plt.ylabel("log(N(mag))")
plt.title("Log(Number of Stars) versus Magnitude")
plt.show()
```

#### Problem 3

```python
datafilestable='stable.dat'
```

```python
stable_color=ascii.read(datasite+datafilestable,header_start=-2,data_start=0)
print(stable_color.colnames)
```

```python
V_stable = stable_color['V']
BV_stable = stable_color['B-V']
```

```python
#how to find true magnitude
#m0 = true magnitude, m = observed magnitude, A is extinction
# m0 = m-A

#A = CE(B-V)

#E(B-V) = 0.25

A = 3.1*0.25

BV_stable_corrected = -(0.25 - BV_stable)

mag_stable_corrected = V_stable - A
```

```python
datafileZAMS02='zams_02.dat'
```

```python
ZAMS02_color=ascii.read(datasite+datafileZAMS02,header_start=-2,data_start=0)
print(ZAMS02_color.colnames)
```

```python
mass_ZAMS02 = ZAMS02_color['mass']
MV_ZAMS02 = ZAMS02_color['M_V']
BV_ZAMS02 = ZAMS02_color['B-V']
```

```python
datafilehyades='hyades.dat'
hyades_color=ascii.read(datasite+datafileZAMS02,header_start=-2,data_start=0)
print(hyades_color.colnames)
```

```python
mass_hyades = hyades_color['mass']
MV_hyades = hyades_color['M_V']
BV_hyades = hyades_color['B-V']

distance_mod_h = 1

distance = 10**((distance_mod_h + 5)/5)

MV_hyades_corrected = (5*np.log10(distance) - 5) + MV_hyades
```

```python
#derive a distance to the stable with ZAMS

distance_mod = 4.5


distance = 10**((distance_mod + 5)/5)
print(distance)
MV_ZAMS02_corrected = (5*np.log10(distance) - 5) + MV_ZAMS02
```

```python
plt.scatter(BV_stable_corrected, mag_stable_corrected, color = 'orange', alpha = .7);
plt.scatter(BV_ZAMS02, MV_ZAMS02_corrected, color = 'green', alpha = .7);


#plt.scatter(BV_hyades, MV_hyades_corrected, color = 'purple', alpha = .7);
plt.gca().invert_yaxis()
plt.xlabel("B-V Color Scale");
plt.ylabel("Apparent Magnitude");
plt.legend(["Stable", "ZAMS_02"])
plt.minorticks_on()

plt.title("Apparent Magnitude versus B-V color Scale")

plt.show();
```
