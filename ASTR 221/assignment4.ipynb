```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as patches
```

```python
dfrv = pd.read_csv('rv.csv')
dftransit = pd.read_csv('transit.csv')
```

```python
periodrv = dfrv['period']
massrv = dfrv['mass']

periodtransit = dftransit['period']
masstransit = dftransit['mass']

periodSS = [88,225,365, 687, 4333,10759,30687,60190]
massSS = [0.0002, 0.0026, 0.0031, 0.0003, 1, 0.3, 0.045,  0.053]
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)

ax1.scatter(np.log10(periodrv), np.log10(massrv), s=0.5, c = "palevioletred", label="RV")
ax1.scatter(np.log10(periodtransit), np.log10(masstransit), s=0.5, c = "green", label="Transits")
ax1.scatter(np.log10(periodSS[0:3]), np.log10(massSS[0:3]), s=7, c = "darkred", alpha = 0.7, label="Rocky Planets")
ax1.scatter(np.log10(periodSS[4:]), np.log10(massSS[4:]), s=7, c = "navy", alpha = 0.7, label="Gas Giants")
plt.xlabel("Log Period (Days)")
plt.ylabel("Log Mass ($M_J$)")
plt.legend()
plt.grid(linestyle = '--', linewidth = 0.5)

plt.title('Transit and Radial Velocity Exoplant Detection Methods')

ax1.add_patch(patches.Rectangle((-.5, -1), # (x,y)
                                1.5, # width
                                2, # height
                                alpha=0.3, facecolor="red", linewidth=3, linestyle='solid'))

ax1.add_patch(patches.Rectangle((1.1, -2), # (x,y)
                                1.4, # width
                                2, # height
                                alpha=0.3, facecolor="cadetblue", linewidth=3, linestyle='solid'))
ax1.add_patch(patches.Rectangle((1.1, 0), # (x,y)
                                2.9, # width
                                1, # height
                                alpha=0.3, facecolor="cadetblue", linewidth=3, linestyle='solid'))

ax1.add_patch(patches.Rectangle((2.6, -2), # (x,y)
                                2, # width
                                1.9, # height
                                alpha=0.3, facecolor="green", linewidth=3, linestyle='solid'))

ax1.add_patch(patches.Rectangle((-.5, -2), # (x,y)
                                1.5, # width
                                .9, # height
                                alpha=0.3, facecolor="yellow", linewidth=3, linestyle='solid'))

ax1.add_patch(patches.Rectangle((-.5, -3.5), # (x,y)
                                5.1, # width
                                1.4, # height
                                alpha=0.3, facecolor="purple", linewidth=3, linestyle='solid'))


ax1.text(-.5, .75, "Hot Jupiters", transform=ax1.transData)
ax1.text(1.1, .75, "Gas Giants", transform=ax1.transData)
ax1.text(3.5, -2, "Ice Giants", transform=ax1.transData)
ax1.text(-.5, -2, "???", transform=ax1.transData)
ax1.text(-.5, -3.5, "Rocky Planets (Earth-like)", transform=ax1.transData)
plt.show()

```
