```python
from astropy.io import ascii
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SDSS=pd.read_csv('SkyServer_SQL_Query.csv')
print(SDSS)
```

```python
#Finding distance to the objects
z = SDSS['z']
v = z * 3e5
d = v / 72 * 1e6
plt.hist(d)
```

```python
Abs_U = 5 + SDSS['dered_u'] - (5 * np.log10(d))
Abs_G = 5 + SDSS['dered_g'] - (5 * np.log10(d))
Abs_R = 5 + SDSS['dered_r'] - (5 * np.log10(d))
Abs_I = 5 + SDSS['dered_i'] - (5 * np.log10(d))
Abs_Z = 5 + SDSS['dered_z'] - (5 * np.log10(d))
```

```python
r_i = Abs_R - Abs_I
g_r = Abs_G - Abs_R
plt.scatter(Abs_G, g_r, s = 1, alpha = 0.5, color = 'purple')
plt.xlim(-23,-16)
plt.ylim(-0.2,1.2)
plt.ylabel(r'Color ($g-r$)')
plt.xlabel(r'Absolute $G$ magnitude')
plt.title('Color - Abs. Magnitude Diagram!')
```

```python
plt.scatter(g_r, r_i, s = .5, alpha  = 0.5, color = 'purple')
plt.xlim(-.2,1.2)
plt.ylim(-0.5,0.8)
plt.xlabel(r'$g-r$ color')
plt.ylabel(r'$r-i$ color')
plt.title('Color verus Color Diagram')
```
