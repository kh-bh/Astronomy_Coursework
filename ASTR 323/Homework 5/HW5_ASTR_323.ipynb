```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.special import gamma,gammainc, gammaincc
```

```python
l_star = 1.5e10
norm_star = 0.05
x = 5e8/l_star
```

```python
alpha_1 = np.linspace(0,-1,200, endpoint=False)
alpha_2 = np.linspace(0,-2,200, endpoint=False)
```

```python
dwarf_num = norm_star*gammainc(alpha_1+1,x)
giant_num = norm_star*gammaincc(alpha_1+1,x)
plt.plot(alpha_1,np.log10(dwarf_num/giant_num))
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$\log(\frac{N_{dwarf}}{N_{giant}})$")
plt.title(r"Log of Total Galaxies ($\frac{dwarf}{giant}$) versus Alpha")
found_x = np.interp(0, np.log10(dwarf_num/giant_num), alpha_1)
print(found_x)
```

```python
dwarf_lum = l_star*norm_star*gammainc(alpha_2+2,x)
giant_lum = l_star*norm_star*gammaincc(alpha_2+2,x)
plt.plot(alpha_2,np.log10(dwarf_lum/giant_lum))
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$\log(\frac{L_{dwarf}}{L_{giant}})$")
plt.title(r"Log of Total Luminosity ($\frac{dwarf}{giant}$) versus Alpha")
plt.xlim(-2)
found_x = np.interp(0, np.log10(dwarf_lum/giant_lum), alpha_2)
print(found_x)
```

```python
lum_den_tot_d = np.interp(-1, alpha_2, dwarf_lum)
lum_den_tot_g = np.interp(-1, alpha_2, giant_lum)
print(lum_den_tot_d+lum_den_tot_g)
```
