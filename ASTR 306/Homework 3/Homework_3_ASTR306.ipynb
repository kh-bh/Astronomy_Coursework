```python
import numpy as np
import matplotlib.pyplot as plt
```

### Question 1

```python
fwhm = 1.2
sigma = fwhm/2.3548
r = np.linspace(0,3,100)
enclosed_light = 1 - np.exp(-r**2) / (2*sigma**2)

plt.plot(r, enclosed_light, color = 'k')
plt.axhline(0.5, color='pink', linestyle='--', label='50% Light')
plt.axhline(0.8, color='blue', linestyle='-.', label='80% Light')
plt.axhline(0.99, color='red', linestyle=':', label='99% Light')
plt.title('Radius needed for Specifics Fraction of Light')
plt.xlabel('Radius (arcsec)')
plt.ylabel('Fraction of Enclosed Light')
plt.legend()
plt.show()
```
